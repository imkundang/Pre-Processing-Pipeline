```python
def data_shape(data):
    # find the shape of data
    shape=data.shape
    rows = data.shape[0]
    col= data.shape[1]
    return {'shape of data':shape, 'number of rows':rows, 'number of features':col }

data_shape(df)
```

```python
def nullcount(data):
    total=(data.isnull().sum()).sort_values(ascending=False)
    percentage=((data.isnull().sum())/(data.isnull().count())*100).sort_values(ascending=False)
    datatype=data.dtypes
    ms=pd.concat([total,percentage,datatype],axis=1,keys=['Total','Percentage','Datatype'])
    ms=ms[ms['Percentage']>0]
    return ms

nullcount(df)
```

```python
def finddtye(data):
    intcount=0
    floatcount=0
    objectcount=0
    for name,datatype in data.dtypes.iteritems():
        if datatype=='float64':
            floatcount=floatcount+1
        elif datatype=='int64':
            intcount=intcount+1
        else:
            objectcount=objectcount+1
    df=pd.DataFrame({'IntegerType':[intcount],'Float Type':[floatcount],'Object Type':[objectcount]})
    df['Total Features']=df['IntegerType']+df['Object Type']+df['Float Type']
    return df
finddtye(df)
```

```python
#let's findout unique value in each of categorical variables
def col_enc(data):
    for col in data.select_dtypes(include=['object']).columns:
        print(col, ':' , len(data[col].unique()), 'labels')
col_enc(df)
```

```python
def dropcolumns(data):
    for i in data.select_dtypes(include=['object']).columns:
        data.drop(i,axis=1,inplace=True)
dropcolumns(df)
```

```python
def filldata(data):
    for name,datatype in data.dtypes.iteritems():
        if datatype=='float64':
            print(name)
            data.fillna(data[name].median(),inplace=True)
        elif datatype=='int64':
            print(name)
            data.fillna(data[name].mean(),inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
def data_split(data):
    X=data.iloc[:,:-1]
    y=data['loss']
    x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=5,test_size=.3)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)

    return x_train,x_test,y_train,y_test
```
